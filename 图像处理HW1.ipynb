```python
from PIL import Image
import cv2 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/home/ponytai1/PycharmProjects/learn_pandas/图像处理HW1/光纤/content/unnormals/unnormals1/defection(1).bmp'
img = cv2.imread(path)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # 转化为灰度图像
```

```python
#plt.subplot(111)
plt.imshow(img)
plt.show()
```

```python
#img2 = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #　转换成PIL RGB通道格式
image = Image.fromarray(cv2.cvtColor(img,cv2.COLOR_GRAY2RGB))
plt.imshow(image)
```

```python
r, b = cv2.threshold(img,200,210,cv2.THRESH_TOZERO_INV)#二值化处理,处理前后背景

img2 = Image.fromarray(cv2.cvtColor(b, cv2.COLOR_GRAY2RGB))#　转换成PIL RGB通道格式
plt.imshow(img2)
```

```python
# 二进制阈值化处理
r, b1 = cv2.threshold(b, 195, 255, cv2.THRESH_BINARY)

img6 = Image.fromarray(cv2.cvtColor(b1, cv2.COLOR_GRAY2RGB))#　转换成PIL RGB通道格式
plt.imshow(img6)
```

```python
kernel = np.ones((10,10), np.uint8)
#图像腐蚀处理
b1 = cv2.erode(b, kernel,200)

img4 = Image.fromarray(cv2.cvtColor(b1, cv2.COLOR_GRAY2RGB))#　转换成PIL RGB通道格式
plt.imshow(img4)
```

```python
#设置卷积核
kernel = np.ones((5,5), np.uint8)
#图像膨胀处理
b = cv2.dilate(b, kernel,1)

img5 = Image.fromarray(cv2.cvtColor(b, cv2.COLOR_GRAY2RGB))#　转换成PIL RGB通道格式
plt.imshow(img5)
```

```python
# 二进制阈值化处理
r, b1 = cv2.threshold(b, 50, 255, cv2.THRESH_BINARY)

img6 = Image.fromarray(cv2.cvtColor(b1, cv2.COLOR_GRAY2RGB))#　转换成PIL RGB通道格式
plt.imshow(img6)
```

```python
mask = b5
image1=cv2.add(img, np.zeros(np.shape(img), dtype=np.uint8), mask=mask)

img2 = Image.fromarray(cv2.cvtColor(image1, cv2.COLOR_GRAY2RGB))#　转换成PIL RGB通道格式
plt.imshow(img2)
```
